```python
data.reset_index(inplace = True)
```

```python
import pycountry_convert as pc
dict_maps={"Cote d'Ivoire":'Africa','Congo (Brazzaville)':'Africa','Congo (Kinshasa)':'Africa','Saint Helena':'Africa','Reunion':'Africa',
           'Midway Islands':'Oceania','Antarctica':'Antarctica','Virgin Islands':'South America','Netherlands Antilles':'South America','Burma':'Asia',
          'East Timor':'Asia','Johnston Atoll':'Oceania','Western Sahara':'Africa','Wake Island':'Oceania','Bolivia (Plurinational State of)': 'South America', 'China, Hong Kong SAR':'Asia', 
         'China, Macao SAR':'Asia', 'D.P.R. of Korea' : 'Asia', 'D.R. of the Congo' : 'Africa', 'Iran (Islamic Republic of)' : 'Asia', "Lao People's DR": 'Asia', 'Micronesia (FS of)' : 'Oceania',
          'Republic of Korea': 'Asia', 'St. Vincent and the Grenadines' : 'South America', 'State of Palestine' : 'Asia', 'U.R. of Tanzania: Mainland': 'Africa', 'Venezuela (Bolivarian Republic of)' : 'South America'} #создаем словарь с отсутствующими странами
def country_to_continent(country_name):
    try:
        country_alpha2 = pc.country_name_to_country_alpha2(country_name)
        country_continent_code = pc.country_alpha2_to_continent_code(country_alpha2)
        country_continent_name = pc.convert_continent_code_to_continent_name(country_continent_code)
        return country_continent_name
    except:
        return dict_maps[country_name]


data['Continents']=data['Country'].map(country_to_continent)
data.head()
```

```python
conts = data.iloc[:,2:].groupby('Continents').mean()
conts
```

```python
fig = plt.figure(figsize=(30,15))
for i in conts.index:
    plt.plot(conts.loc[f'{i}',:], label = f'{i}')
plt.title('Mean per Capita GDP of Continents in US dollar', size = 40)
plt.xlabel('Year', size = 20)
plt.ylabel('Dollars', size = 20)
plt.legend()
```

###### Видим небольшую просадку по всем континентам в начале двухтысячных из-за пузыря доткомов. Также тяжело не заметить "галки" в 2008-2009 из-за мирового финансового кризиса и в 2020-2021 из-за ковида

##### Посчитаем долю каждого континента в мировом ввп

```python
plt.pie(conts['2022'],labels = conts.index,autopct='%1.1f%%')
```

##### Рассмотрим детально уровень ввп на душу населения на фоновой картограмме

```python
fig = px.choropleth(data,
                    locations='Country', locationmode='country names',
                    color = '2022',hover_name="Country",
                    color_continuous_scale='Viridis_r')
fig.update_layout(margin={'r':0,'t':0,'l':0,'b':0}, coloraxis_colorbar=dict(
    title = 'GDP - 2022',
    ticks = 'outside',
    tickvals = [0,50000,100000,150000,200000,250000],
    dtick = 12))              
fig.show()
```

```python
data1 = data.groupby('Continents')[['Country','2022','1970']].apply(lambda x : x.nlargest(5,['2022','1970'])).reset_index(level = 1, drop=True)
data1
```

##### Рассмотрим, насколько сильно за 52 года изменился ввп ведущих стран континентов

```python
fig = plt.figure(figsize=(60,30))
ax = sns.barplot(data = data1,x = 'Country', y = '2022', hue = data1.index)
sns.barplot(data = data1,x = 'Country', y = '1970',ax = ax, width = 0.1, label  = '1970', color = 'maroon')
plt.title('Top 5 countries in each continent with the highest GDP in 2022 compared to 1970 ', size = 40)
plt.legend()
```
